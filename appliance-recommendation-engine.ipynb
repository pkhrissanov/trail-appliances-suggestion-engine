# There are the actual recommendations. It works by picking a specific product index, and then each model spits out a certain number of recommendations.

# Cosine Similarity

```python
import pandas as pd
data = pd.read_csv("data/final_data.csv")
data_readable = pd.read_csv("data/ranges.csv")
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_matrix = cosine_similarity(data)

product_index = 0
similar_products = similarity_matrix[product_index].argsort()[::-1][1:5]
print("Looking at product: \n", data_readable.iloc[product_index])
for i in similar_products:
    print("\n",data_readable.iloc[i])
```

# Nearest Neighbours

```python
from sklearn.neighbors import NearestNeighbors

model = NearestNeighbors(n_neighbors=5, metric="cosine")
model.fit(data)

distances, indices = model.kneighbors(data.iloc[product_index].values.reshape(1, -1))
print(indices)
```

# K means clustering

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=8, random_state=42)
data["cluster"] = kmeans.fit_predict(data)
data
```
