# Appliance Recommendation Engine

## Goal

This project explores multiple machine learning techniques for recommending similar appliances.

Given a selected appliance, the engine returns the most similar products based on characteristics such as:

- Brand
- Appliance type
- Width
- Price
- Features
- Rating
- Smart capabilities

The purpose of this project is to investigate different recommendation algorithms and eventually build a context-aware recommendation system similar to those used in e-commerce.

# 1. Cosine Similarity

Cosine similarity compares every appliance to every other appliance by treating each product as a feature vector.

Products with the highest cosine similarity are considered the most alike.

This serves as the baseline recommendation algorithm.

```python
import pandas as pd
data = pd.read_csv("data/final_data.csv")
data_readable = pd.read_csv("data/ranges.csv")
product_index = 23
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_matrix = cosine_similarity(data)

similar_products = similarity_matrix[product_index].argsort()[::-1][1:5]
print("Base product: \n", data_readable.iloc[product_index], "\n")
print(data_readable.iloc[similar_products])
```
